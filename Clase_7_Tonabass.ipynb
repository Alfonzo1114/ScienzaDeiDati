```python
class cola:
    def __init__(self, capacidad = 100):
        self.datos = []
        self.capacidad = capacidad
    
    def enqueue(self, n):
        if len(self.datos) < self.capacidad:
            self.datos.insert(0, n)

    def dequeue(self):
        return self.datos.pop()

    def empty(self):
        return self.datos == []
```
